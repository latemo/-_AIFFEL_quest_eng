```python
import math

while True:
    try:
        num1 = int(input("첫 정수: "))
        num2 = int(input("두 정수: "))
    except ValueError:
        print("정수만 입력하세요.")
        continue

    op = input("연산자(+,-,*,/,**): ")

    if op == "+":
        print("결과:", num1 + num2)
    elif op == "-":
        print("결과:", num1 - num2)
    elif op == "*":
        print("결과:", num1 * num2)
    elif op == "/":
        if num2 == 0:
            print("0으로 나눌 수 없습니다.")
        else:
            print("결과:", num1 / num2)
    elif op == "**":
        print("결과:", math.pow(num1, num2))
    else:
        print("지원하지 않는 연산자입니다.")

    again = input("계속? (y/n): ")
    if again != "y":
        break
```
