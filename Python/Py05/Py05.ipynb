```python
import random

class Account:
    # Q2: 클래스 변수 - 전체 생성된 계좌 수
    account_count = 0

    def __init__(self, owner, balance=0):
        # Q1: 속성 초기화 및 계좌번호 자동 생성
        self.bank_name = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = f"{random.randint(0, 999):03d}-{random.randint(0, 99):02d}-{random.randint(0, 999999):06d}"
        
        # 이자 지급용 카운터
        self.deposit_count = 0
        
        # Q10: 입출금 내역 저장 리스트
        self.deposit_log = []
        self.withdraw_log = []
        
        # 계좌 생성 시 카운트 증가
        Account.account_count += 1

    # Q3: 전체 계좌 수 반환 (클래스 메서드)
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4 + Q7: 입금 메서드 (1원 이상, 5회마다 1% 이자)
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_log.append(amount)
            print(f"[{self.owner}] {amount:,}원 입금 완료")
            
            # 5회 입금 시 1% 이자 지급
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"🎉 [{self.owner}] 5회 입금 달성! 1% 이자({interest:,}원) 지급 완료.")
        else:
            print("[입금 실패] 입금액은 1원 이상이어야 합니다.")

    # Q5: 출금 메서드 (잔고 이상 또는 동일 금액 출금 불가)
    def withdraw(self, amount):
        if amount >= self.balance:
            print(f"[{self.owner} 출금 실패] 잔고가 부족하거나 전액 출금은 불가합니다.")
        else:
            self.balance -= amount
            self.withdraw_log.append(amount)
            print(f"[{self.owner}] {amount:,}원 출금 완료")

    # Q6: 계좌 정보 출력 (잔고 천 단위 쉼표)
    def display_info(self):
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    # Q10: 입금 내역 출력
    def deposit_history(self):
        print(f"\n--- [{self.owner}] 입금 내역 ---")
        if not self.deposit_log:
            print("입금 내역이 없습니다.")
            return
        for amount in self.deposit_log:
            print(f"+ {amount:,}원")

    # Q10: 출금 내역 출력
    def withdraw_history(self):
        print(f"\n--- [{self.owner}] 출금 내역 ---")
        if not self.withdraw_log:
            print("출금 내역이 없습니다.")
            return
        for amount in self.withdraw_log:
            print(f"- {amount:,}원")


# ==========================================
# 실행 및 테스트 (Q8, Q9, Q10 포함)
# ==========================================
if __name__ == "__main__":
    # Q8: 여러 계좌 생성 및 리스트 저장
    acc1 = Account("고길동", 10000)
    acc2 = Account("금잔디", 50000)
    acc3 = Account("홍금보", 1200000)   # 100만원 이상
    acc4 = Account("이미자", 2500000)   # 100만원 이상

    account_list = [acc1, acc2, acc3, acc4]

    print(f"▶ 총 개설된 계좌 수: {Account.get_account_num()}개\n")

    # 전체 계좌 정보 출력
    print("--- 전체 계좌 정보 ---")
    for acc in account_list:
        acc.display_info()

    # 이자 지급 테스트 (고길동 계좌)
    print("\n--- 이자 지급 테스트 (고길동) ---")
    for _ in range(5):
        acc1.deposit(1000)
    acc1.display_info()

    # 거래 내역 출력
    acc1.deposit_history()
    acc1.withdraw_history()

    # Q9: 잔고 100만원 이상 고객만 출력
    print("\n--- 잔고 100만원 이상 고객 정보 ---")
    for acc in account_list:
        if acc.balance >= 1000000:
            acc.display_info()
```
